```python
import random
number=random.randint(1,10)
k=3
win=int
print('Угадайте число от 1 до 10')

while k>0:
  print('количество попыток=',k)
  userInput=int(input('введите число '))
  
  if userInput==number:
    k=0
    win=1
  else:
    k=k-1
    if userInput>number:
      print('Заданное число меньше')
    else:
      print('Заданное число больше')
if win==1:
  print('Вы угадали!')
else:
  print('Попытки закончились')

  
 
```
