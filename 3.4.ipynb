```python
print('квадратное уравнение формата  ax^2+bx+c=0')
a=float(input('введите а '))
b=float(input('введите b '))
c=float(input('введите c '))
D=b**2-4*a*c
if D<0:
  print('Дискриминант отрицательный, решений НЕТ')
if D==0:
  x=-b/(2*a)
  print('x=', x)
if D>0:
  x1=(-b+D**0.5)/(2*a)
  x2=(-b-D**0.5)/(2*a)
  print('корни уравнения:')
  print('x1=',x1,'  x2=',x2)
```
